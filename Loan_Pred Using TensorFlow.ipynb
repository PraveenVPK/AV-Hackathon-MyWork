```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
import numpy as np
import pandas as pd
# fix random seed for reproducibility
numpy.random.seed(7)
# load pima indians dataset
#dataset = pd.read_csv("train_u6lujuX_CVtuZ9i.csv")
# split into input (X) and output (Y) variables
print(dataset.shape)
#X = dataset.iloc[:,0:13]
#Y = dataset.iloc[:,13]

X,Y  = dataset.drop(['Loan_Status','Loan_ID','Education','Self_Employed','Dependents','Gender','Married'],axis=1), dataset['Loan_Status']

# create model
model = Sequential()
#model.add(activation(tf.nn.softmax))
model.add(Dense(25, input_dim=5, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(40, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
from keras.callbacks import TensorBoard

tensorboard = TensorBoard(log_dir='tensor_simple', histogram_freq=0,
                          write_graph=True, write_images=True)
#print(X)
# Fit the model
model.fit(X, Y, epochs=350, batch_size=80,shuffle=True,callbacks=[tensorboard])
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
